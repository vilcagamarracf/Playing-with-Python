### Lectura de archivo raster (`.tiff`)

Haremos uso de la librer칤a [Rasterio](https://rasterio.readthedocs.io/en/latest/) para la lectura y visualizaci칩n de archivos `.tiff`.

````python
# Importar librer칤as

import rasterio 
from rasterio.plot import show

rasterio.__version__
````

````python
# Listado de m칠todos disponibles en rasterio

for i, value in enumerate(dir(rasterio)):
    print(i, value)
````

````python
# Listado de archivos en carpeta
os.listdir() 
````

````python
# Path
image_file = "chincha_s2_4_bands.tif"

# Opening the image
sat_data = rasterio.open(image_file)

# Printing the metadata
from pprint import pprint
print('Metadata:')
pprint(sat_data.meta)

print('Image Information:\n', sat_data)    # Image Information
print('Bands count:\n', sat_data.count)    # bands number
print('Band indexes:\n', sat_data.indexes) # sequence of band indexes R-G-B-Nir (4 bands)
print('Shape:\n', sat_data.shape)          # shape: (height, width) 
print('Bounds:\n', sat_data.bounds)        # limites
````

Una imagen satelital es una imagen digital obtenida mediante el proceso de conversi칩n de la energ칤a electromagn칠tica registrada mediante sensores abordo de un sat칠lite a valores de pixel. Por ello, una imagen digital no es m치s que una malla de pixeles donde cada pixel tiene un valor el cual representa la informaci칩n de reflectancia reflejada por ese mismo punto en superficie y captada por el sensor en el espacio.

````python
# Representaci칩n de una imagen satelital en forma de tipo array
# Nota: Son 4 bandas -> 4 matrices 2D : 1 matriz 3D
sat_data.read()
````

````python
# Asignar cada banda en un variable por separado (visualizaci칩n)
b, g, r, nir = sat_data.read()
b, g, r, nir
````

````python
# Informaci칩n de cada banda (dimensiones y valor min-max)
for i in range(sat_data.count):
    print('Banda {a}: height-width {b} y min-max {c}'.format(
      a = i+1, 
      b = sat_data.read(i+1).shape,
      c = (sat_data.read(i+1).min(), sat_data.read(i+1).max())
      )
  )
````

### Visualizaci칩n de imagen

Se puede visualizar las im치genes mediante dos formas:
- Usando matplotlib.pyplot
- Usando rasterio

#### Visualizaci칩n usando Matplotlib

Posee por defecto la visualizaci칩n de im치genes (`.png`, `.jpg`, `.gif`, y m치s) de tipo `uint-8` m치ximo.

````python
import matplotlib.pyplot as plt
%matplotlib inline

# Visualizaci칩n
plt.figure(1,figsize=(20,5))
for i in range(1,4+1):
    # 1: Blue, 2: Green, 3: Red y 4: NIR 
    plt.subplot(1,4,i), plt.imshow(sat_data.read(i), cmap='gray'), plt.title('Banda '+str(i));
    # plt.subplot(122), plt.imshow(r, cmap='gray') # cmap= https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
````

Normalmente cuando uno hace una visualizaci칩n con matplotlib.pyplot (com칰nmente asignado como `plt`), se producir치 un mensaje junto a la visualizaci칩n. Dicho mensaje es "ocultado" usando el m칠todo `plt.show()`. 

Dependiendo de si querer o no que se muestre, yo uso `%matplotlib inline` (usar solo una vez en la libreta) ya que permite usar un `;` al final del c칩digo para visualizar y reemplazar la funci칩n del `plt.show()`.

#### Visualizaci칩n usando Rasterio

Permite la visualizaci칩n de archivos `geotiff` (aquellos que poseen metadata adjunto con la imagen) de cualquier tipo y a su vez asigna la informaci칩n de la georeferenciaci칩n (de la metadata) a los ejes de cada imagen.

````python
# Visualizaci칩n b치sica

# Read and show the image
sen2_arr = sat_data.read()

fig, ax = plt.subplots(figsize=[10,10], ncols=1, nrows=1)
show(sat_data, cmap='gray',ax=ax) # de rasterio.plot
#ax.set_axis_off() # para omitir ejes
plt.show()
````

**Visualizaci칩n con ajuste de histograma del 5-95%**

Modificaci칩n del histograma:
```
vmin, vmax = np.nanpercentile(sen2_arr, (5,95))  # 5-95% contrast stretch
```
donde se asigna el valor m칤nimo a `vmin` y el valor m치ximo a `vmax`, los cuales se usar치n como par치metros en el m칠todo `show` de rasterio.plot.

````python
# Visualizaci칩n con ajuste de histograma

# Lectura de imagen
sen2_arr = sat_data.read()

import numpy as np
vmin, vmax = np.nanpercentile(sen2_arr, (5,95))  # 5-95% contrast stretch

fig, ax = plt.subplots(ncols=1, nrows=1, figsize=[10,10]) 
show(sat_data,cmap='gray', vmin=vmin, vmax=vmax, ax=ax); # de rasterio.plot
#ax.set_axis_off() # para omitir los ejes 
````

N칩tese que los ejes se encuentan en coordenadas geogr치ficas ya que rasterio hace uso de la metadata de la imagen.

**Mejor visualizaci칩n: Color natural**

````python
# Mejor visualizaci칩n - Combinaci칩n Red-Green-Blue

from rasterio.plot import show, adjust_band
imgdata = np.array([adjust_band(sat_data.read(i)) for i in (3,2,1)]) # RGB

fig, ax = plt.subplots(1,1,figsize=(10,10)) # o: fig, ax = plt.subplots(nrows=1, ncols=1, figsize=[21,21])
show(imgdata*2, ax=ax, title='RGB * 2 Brightness'); # Factor 2 es usado para incrementar el brillo

# Saldr치 una mensaje ya que se normaliza las bandas (valores de 0-1) 
````

En 칠ste ejemplo se ha hecho:
```
imgdata = np.array([adjust_band(sat_data.read(i)) for i in (3,2,1)])
```
donde se est치 aplicando el m칠todo `adjust_band` de rasterio.plot a cada banda para su normalizaci칩n haciendo que los ejes muestren informaci칩n de la cantidad de pixeles en la imagen, perdi칠ndose la ubicaci칩n por coordenadas geogr치ficas.

### Visualizaci칩n de histograma

Mediante el uso de histogramas es posible observar la cantidad de pixeles en cada banda (imagen satelital) o canal (imagen digital).

#### Usando matplotlib

````python
# Usando el m칠todo .indexes en un loop 
print(sat_data.indexes)
print(type(sat_data.indexes))

for i in sat_data.indexes:
    print(i)
````

````python
# Visualizaci칩n de histograma por bandas
plt.figure(1, figsize=(20,5)) 
for i in sat_data.indexes:
    plt.subplot(1,4,i)
    plt.title('Band '+str(i))
    plt.hist(np.ravel(sat_data.read(i)), bins=256, range=(0,7000), histtype='step') # histtype='stepfilled'
    
````

#### Usando rasterio

El m칠todo `show_hist` de Rasterio permite visualizar un histograma general con informaci칩n de las bandas de la imagen satelital.

````python
# Histograma

from rasterio.plot import show_hist

plt.figure(1, figsize=(15,7))
show_hist(sat_data, bins=1000, lw=0.0, alpha=0.2, histtype='stepfilled', title='Histogram')
````
